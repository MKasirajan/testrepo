Question 2: Use Webscraping to Extract Tesla Revenue Data

Display the last five rows of the tesla_revenue dataframe using the tail function.

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the URL
url = "https://finance.yahoo.com/quote/TSLA/history?p=TSLA"
response = requests.get(url)

# Parse the HTML content of the page
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the revenue data
table = soup.find('table', {'class': 'W(100%) M(0)'})

# Extract the rows of the table
rows = table.find_all('tr')

# Create a list to store the data
data = []

# Iterate over the rows and extract the data from each cell
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) # Get rid of empty values

# Convert the data list into a Pandas dataframe
tesla_revenue = pd.DataFrame(data, columns=['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])

# Display the last five rows of the dataframe using the tail function
print(tesla_revenue.tail())
```

Question 5: Plot Tesla Stock Graph

Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a Pandas dataframe from the given data
data = [
    ['2022-12-19', '766.00', '777.00', '761.00', '765.25', '765.25', '2401400'],
    ['2022-12-20', '765.50', '776.00', '754.00', '769.50', '769.50', '3104200'],
    ['2022-12-21', '770.50', '772.50', '760.00', '761.75', '761.75', '2399600'],
    ['2022-12-22', '762.00', '765.50', '756.50', '762.00', '762.00', '1509500'],
    ['2022-12-23', '761.00', '761.50', '756.00', '758.50', '758.50', '1144100'],
]
tesla_revenue = pd.DataFrame(data, columns=['Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume'])

# Convert the 'Date' column to a datetime object
tesla_revenue['Date'] = pd.to_datetime(tesla_revenue['Date'])

# Set the 'Date' column as the index of the dataframe
tesla_revenue.set_index('Date', inplace=True)

# Make the graph
def make_graph(dataframe):
    plt.plot(dataframe['Close'])
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.title('Tesla Stock Data')
    plt.show()

make_graph(tesla_revenue)
```
